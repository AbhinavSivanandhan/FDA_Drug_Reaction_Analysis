```python
import dask.dataframe as dd

# Load the Parquet dataset
parquet_path = "parquet_database"
df = dd.read_parquet(parquet_path)

# Inspect the dataset
print(df.head())  # Displays the first few rows
print(df.dtypes)  # Check column data types

# Total records (lazy computation)
total_records = df.shape[0].compute()
print(f"Total records: {total_records}")
```

```python
import os

for root, dirs, files in os.walk("parquet_database"):
    print(f"Directory: {root}")
    for file in files:
        if file.endswith(".parquet"):
            print(f"  File: {file}")
```

```python
import pyarrow.parquet as pq
import pyarrow as pa

# Consolidate the dataset into a single Parquet file
dataset = pq.ParquetDataset("parquet_database", use_legacy_dataset=False)
table = dataset.read()

# Write to a new consolidated Parquet file
pq.write_table(table, "repaired_dataset.parquet")
```
